```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Завантаження зображень
image1 = cv2.imread('chess_1.png')
image2 = cv2.imread('chess_2.png')

# Визначення точок відповідності
points1 = np.array([[211, 80], [448, 74], [215, 218], [442, 231]], dtype=np.float32)
points2 = np.array([[223, 311], [219, 66], [381, 346], [472, 117]], dtype=np.float32)

# Знаходження гомографії
h, _ = cv2.findHomography(points2, points1, method=cv2.RANSAC)

# Застосування гомографії на друге зображення
transformed_image = cv2.warpPerspective(image2, h, (image1.shape[1], image1.shape[0]))

# Відображення результатів
plt.figure(figsize=(10, 6))

# Показ першого зображення
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Chess 1')
plt.axis('off')

# Показ другого зображення
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Chess 2')
plt.axis('off')

# Показ перетвореного зображення
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB))
plt.title('Transformed Chess 2')
plt.axis('off')

plt.tight_layout()
plt.show()

```
